# Calculating the complexity of GNFS algorithm 

```python
def gnfs_complexity_base2(n):
    logn = n*math.log(2)
    #print(f'n={n},logn = {logn}')
    c=(64/9)**(1/3)
    L = math.log(math.exp((c+1)*(logn**(1/3))*(math.log(logn)**(2/3))),2)
    return L



def gnfs_complexity_v1(N):
    lnN = math.log(N)
    #print(f'n={N},logn = {lnN}')
    lnlnN = math.log(lnN)
    c = (64.0 / 9.0) ** (1.0 / 3.0)
    complexity = math.exp( (c+1) * (lnN ** (1.0 / 3.0)) * (lnlnN ** (2.0 / 3.0)) )
    return complexity

# Exemplo de uso
print(f'O(n=45) = {gnfs_complexity_base2(n=45)}')
print(f'O(n=40) = {gnfs_complexity_base2(n=40)}') 
print(f'O(n=512) = {gnfs_complexity_base2(n=512)}') 
print(f'O(n=1024) = {gnfs_complexity_base2(n=1024)}') 
print(f'O(n=512) = {gnfs_complexity_v1(2**512)}') 
print(f'log_2(O(n=512)) = {math.log(gnfs_complexity_v1(2**512),2)}') 
print("\n\n")
for i in range(80,100):
    print(f'O(n={i}) = {gnfs_complexity_base2(n=i)}, n/2-3={i/2-3}')

print("\n\nTable of GNFS complexity vs n\n")
for n in [128,256,512,1024,2048,4096]:
    result = [n, f'{gnfs_complexity_base2(n=n):.8}', f'{gnfs_complexity_base2(n=n)-n-1:.8}']
    print('\t'.join(map(str,result)))
    
```
